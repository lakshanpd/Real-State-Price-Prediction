```python
def count_outliers(df, col, threshold = 1.5):
    counter = 0
    q1, q2, q3 = df[col].quantile([0.25,0.5,0.75]).to_list()
    iqr = q3 - q1
    min = q1 - threshold * iqr
    max = q3 + threshold * iqr

    for point in df[col].to_list():
        if point < min or point > max :
            counter += 1
            
    return counter
```

```python
def clean_outliers(df, col, threshold = 1.5):
    indexes = []
    q1, q2, q3 = df[col].quantile([0.25,0.5,0.75]).to_list()
    iqr = q3 - q1
    min = q1 - threshold * iqr
    max = q3 + threshold * iqr
    arr = df[col].to_list()

    for i in range(len(arr)):
        if arr[i] >= min and arr[i] <= max :
            indexes.append(i)
            
    return df.loc[indexes].reset_index()
```

Probably, ry_renovated = 0 when the renovated has not been done. So, applying the built year for those 0 rows may be more suitable for that.

```python
def yr_renovated_feature_engineering(row, col1, col2):
    if row[col2] == 0:
        row[col2] = row[col1]
    return row
```

**Building data preprocessing pipeline**

```python
df_1.head()
```

```python
def pipeline(model, scaler, encoder, bedrooms, bathrooms, sqft_living, sqft_lot, floors, view, sqft_above, yr_built, yr_renovated, city):
    vector_1 = np.array([bedrooms, bathrooms, floors, view])
    vector_2 = scaler.transform([[sqft_living, sqft_lot, sqft_above, yr_built, yr_renovated]]).flatten()
    cities = encoder.categories_[0]
    vector_3 = np.zeros(len(cities))
    city_index = np.where(cities == city)[0]
    vector_3[city_index] = 1
    vector = np.concatenate((vector_1[:2], vector_2[:2], vector_1[2:], vector_2[2:], vector_3))
    vector = vector.reshape(1, -1)
    return np.round(model.predict(vector), 2)
```

```python
with open('pipeline.pkl', 'wb') as file:
    pickle.dump(pipeline, file)
```
